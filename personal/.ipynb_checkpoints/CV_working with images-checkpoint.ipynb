# Computer Vision - Working with images 

```python
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
img = io.imread("../datasets/girl.jpg")
plt.imshow(img)
plt.show()
```

```python
img.shape
```

# Extracting all 3 channels. i.e: Red, Green, Blue

```python
red = img[:,:,0]
green = img[:,:,1]
blue = img[:,:,2]
```

```python
red
```

# Extract the various color image insted of array matrix 

```python
plt.imshow(red, cmap='Reds')
plt.show()
```

```python
plt.imshow(green, cmap = "Greens")
plt.show()
```

```python
plt.imshow(blue, cmap = 'Blues')
plt.show()
```

# Affine Transformation

```python
import tensorflow as tf
```
