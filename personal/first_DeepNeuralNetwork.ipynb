```python
import pandas as pd
import numpy as np
import keras 
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
data = loadtxt('../datasets/diabetes_noHeader.csv', delimiter=',')
data
```

```python
x = data[:,0:8]
```

```python
x
```

```python
y = data[:,8]
```

```python
y
```

```python
x_df = pd.DataFrame(x)
x_df
```

```python
y_df = pd.DataFrame(y, columns=['Y'])
y_df
```

# 2. Create a sequential model# 3. add layers

```python
from keras.models import Sequential
model = Sequential()
```

# 3. add layers

```python
model.add(Dense(units=12, input_dim=8 ,activation='relu',))
model.add(Dense(units=8, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

# 4. compile model

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

# 5. fit model with training dataset

```python
model.fit(x,y,epochs=150, batch_size=10)
```

# to check the less loss and high accuuracy 

```python
loss, accuracy = model.evaluate(x,y) 
```

# 6. Predict values 

```python
pred = model.predict(x)
rounded = [round(x[0]) for x in pred]
rounded
```

```python
pred_value = pd.DataFrame(rounded, columns=['y_Pred'])
```
